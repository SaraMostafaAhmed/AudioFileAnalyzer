```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
life = pd.read_csv('/content/Life_Satisfaction.csv')
gdp=pd.read_excel('/content/GDPPerCapitaexc.xlsx')
```

```python
life.head()
```

```python
gdp.head()
```

```python
life.drop_duplicates(inplace=True)
gdp.drop_duplicates(inplace=True)
```

```python
data=life.merge(gdp,left_on='Country',right_on='GDP per capita, current prices\n (U.S. dollars per capita)',how='inner')
data
```

```python
data.drop_duplicates(inplace=True,subset='Country')
```

```python
data
```

```python
data.shape
```

```python
data.head()
```

```python
df=data[['Country','Value',2023]].copy()
df.head()
```

```python
df.set_index('Country',inplace=True)
```

```python
df
```

```python
target='Value'
x=df.drop(columns=[target])
y=df[target]
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,shuffle=True
)
```

```python
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
r_seq=regressor.score(X_train,y_train)
r_seq_test=regressor.score(X_test,y_test)
```

```python
print(r_seq)
print(r_seq_test)
```
